# Importing libs

```python
import bar_chart_race as bcr
import pandas as pd
```

# File read and DataFrame

```python
df = pd.read_csv('owid-covid-data.csv')
```

```python
df_new = df[['location','date','total_cases']]
```

```python
df_new.head(100)
```

# Pivot table for chart race function

```python
df_new = df_new.pivot_table(index="date", columns="location", values="total_cases")
```

```python
df_new.fillna(0, inplace=True)
```

```python
df_new.shape
```

```python
df_new.head()
```

```python
cols =['Italy','China','Australia','Brazil','India','United Kingdom','Pakistan','Spain','Canada','New Zealand']
#You can change the countries, increase or decrease with manually
```

```python
data_deaths = df_new[cols]
data_deaths
```

# Saving chart as a mp4 file

```python
bcr.bar_chart_race(df=data_deaths , filename = "temp.mp4", figsize= (3.5,3) , title="Covid Stats")
```

# Saving chart as a html file

```python
bcr_html = bcr.bar_chart_race(df=data_deaths, filename="temp.html")
```
